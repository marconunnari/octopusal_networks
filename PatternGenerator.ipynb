# Pattern generator

A small utility to generate pattern for the beach. You can wake up an octopus by clicking on it. You can also hold down the mouse and hover on the beach to wake up more octopuses faster. To make it work you should run all the cell below and return on the output of the cell with the html code.

```python
%%HTML
<div id="root" style="text-align:center">
</div>
```

```python
%%javascript

function init(size) {
    var root = document.getElementById('root');  
    root.innerHTML = ''

    var beach = document.createElement('div');
    beach.style.userSelect = "none";
    root.appendChild(beach);

    for (var x = 0; x < size; x++) {
        for (var y = 0; y < size; y++) {
            var octopus = document.createElement('span');
            octopus.style.cursor = 'pointer'
            octopus.innerHTML = '🐙';
            octopus.setAttribute('data-awake', false);
            octopus.setAttribute('data-x', x);
            octopus.setAttribute('data-y', y);
            octopus.addEventListener('mouseover', handleHover);
            octopus.addEventListener('mousedown', toggleOctopus);
            beach.appendChild(octopus);
        }
        beach.appendChild(document.createElement('br'));
    }

    var mouseDown = 0;
    document.body.onmousedown = function() { 
        mouseDown = 1;
    }
    document.body.onmouseup = function() {
        mouseDown = 0;
    }

    function handleHover(event) {
        if (!mouseDown) {
            return false;
        }
        toggleOctopus(event)
    }

    function toggleOctopus(event) {
        var octopus = event.target;
        if (octopus.innerHTML == '🐙') {
            octopus.innerHTML = '💡';
            octopus.setAttribute('data-awake', true);
        } else {
            octopus.innerHTML = '🐙';
            octopus.setAttribute('data-awake', false);
        }
    }

    var generateButton = document.createElement('button');
    generateButton.innerHTML = 'Generate'
    generateButton.addEventListener('click', handleGeneration)
    root.appendChild(generateButton);
    function handleGeneration() {
        var where = ''
        for (var octopus of beach.childNodes) {
            if (octopus.getAttribute('data-awake') == 'true') {
                var x = octopus.getAttribute('data-x');
                var y = octopus.getAttribute('data-y');
                where += `(${x}, ${y}), `
            }
        }
        var whereDiv = document.createElement('div');
        whereDiv.innerHTML = where;
        root.appendChild(whereDiv);
    }

    var resetButton = document.createElement('button');
    resetButton.innerHTML = 'Reset'
    resetButton.style.marginLeft = '20px'    
    resetButton.addEventListener('click', () => init(size))
    root.appendChild(resetButton);   
}


init(28)
```
